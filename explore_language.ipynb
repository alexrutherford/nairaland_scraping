```python
def clean(s):
    s=re.sub(r'\?|\.|\,|\:|\;|\"|\&|\@|\!|\*|\=|\+|\-|\_|\n|\t|\/|\\',' ',s.lower())
    return [w for w in s.split() if not w in stopWords]
```

```python
conn = pymysql.connect(host='localhost',user='root',passwd='pulsetini', db='nairaland',charset='utf8')
cur = conn.cursor()
cur2 = conn.cursor()

table='posts2'
table='comments2'

cur2.execute('select content from '+table)

langCounter=collections.Counter()
vals=[]
lengths=[]
probs=[]
langVals=collections.defaultdict(list)

for r in cur2:
    l=langid.classify(r[0])
    langCounter[l[0]]+=1
    vals.append(l[1])
    lengths.append(len(r[0].split()))
    langVals[l[0]].append(l[1])
```

```python
df=pd.DataFrame(data={'vals':vals,'lengths':lengths})
```

```python
fig,ax=plt.subplots()
n,bins,patches=ax.hist(df.vals,bins=40,log=True,histtype="stepfilled",alpha=0.7)
plt.xlabel('Language Classification Probability')
plt.ylabel('Post Count')
plt.ylim(1,10**6)
plt.xticks([0.2+i*0.2 for i in range(5)])

ax2=ax.twinx()
width=(bins[1]-bins[0])/2.0
ax2.semilogy([b+width for b in bins[0:-1]],np.cumsum(n)/np.sum(n),color='k')
plt.yticks([10**-i for i in range(4)])
plt.ylim(10**-2,1)
plt.ylabel('Cumulative Probability')
plt.tight_layout()
plt.savefig('post_lang_class_dist.png',dpi=200)
```

```python
contentDf=pd.read_sql_query('select content from posts2',conn)
```

```python
contentDf['cleaned']=contentDf.content.apply(clean)
```

```python
dictionary = gensim.corpora.Dictionary(contentDf.cleaned)
```

```python
corpus = [dictionary.doc2bow(text) for text in contentDf.cleaned]
```

```python
tfidf = gensim.models.TfidfModel(corpus)
```

```python
corpus_tfidf = tfidf[corpus]
```

```python
lsi = gensim.models.LsiModel(corpus_tfidf, id2word=dictionary, num_topics=10)
```

```python
corpus_lsi = lsi[corpus_tfidf]
```
